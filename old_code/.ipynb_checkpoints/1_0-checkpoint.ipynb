# Models

## Llama

```python
# BitsAndBytes-Konfiguration für 4-bit
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16,
)

# LLaMA 2 7B laden – quantisiert
try:
    model_id = "meta-llama/Llama-2-7b-hf"
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Tokenizer und Modell laden
    llama_tokenizer = AutoTokenizer.from_pretrained(model_id)
    llama_model = AutoModelForCausalLM.from_pretrained(
        model_id,
        quantization_config=bnb_config,
        device_map="auto"  # Automatische GPU-Nutzung
    )

    # Prompt definieren
    prompt = (
        f"Statement: \"{text}\"\n"
        f"Is this statement true or false?\n"
        f"Answer with only one word: True or False.\n"
        f"Answer:"
    )

    # Eingaben vorbereiten
    inputs = llama_tokenizer(prompt, return_tensors="pt").to(device)

    # Ausgabe generieren
    outputs = llama_model.generate(**inputs, max_new_tokens=1)
    print("LLaMA 2 7B Output:", llama_tokenizer.decode(outputs[0], skip_special_tokens=True))

except Exception as e:
    print("LLaMA 2 7B Modell nicht geladen:", e)
```

```python
# Quantisierungs-Config (4-bit)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

# 1. LLaMA 3.1 8B Instruct (4-bit geladen auf GPU)
try:
    model_id = "meta-llama/Llama-3.1-8B-Instruct"
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Lade Tokenizer & Modell mit Quantisierung
    llama_tokenizer = AutoTokenizer.from_pretrained(model_id)
    llama_model = AutoModelForCausalLM.from_pretrained(
        model_id,
        quantization_config=bnb_config,
        device_map="auto"  # wichtig für große Modelle
    )

    # Prompt vorbereiten
    prompt = (
        f"Statement: \"{text}\"\n"
        f"Is this statement true or false?\n"
        f"Answer with only one word: True or False.\n"
        f"Answer:"
    )

    # Tokenisieren und auf CUDA verschieben
    inputs = llama_tokenizer(prompt, return_tensors="pt").to(device)
    
    # Antwort generieren
    outputs = llama_model.generate(**inputs, max_new_tokens=1)
    print("LLaMA 3.1 8B Output:", llama_tokenizer.decode(outputs[0], skip_special_tokens=True))

except Exception as e:
    print("LLaMA 3.1 8B Modell nicht geladen:", e)
```

## Deepseek

```python
# 4-Bit Konfiguration
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16
)

# Modell laden mit CPU/GPU-Offloading
try:
    model_id = "deepseek-ai/deepseek-llm-7b-base"
    deepseek_tokenizer = AutoTokenizer.from_pretrained(model_id)

    deepseek_model = AutoModelForCausalLM.from_pretrained(
        model_id,
        quantization_config=bnb_config,
        device_map="auto",  # wichtig für Offloading
        offload_folder="offload",  # temporärer CPU-Speicherort
    )

    # Prompt
    prompt = (
        f"Statement: \"{text}\"\n"
        f"Is this statement true or false?\n"
        f"Answer with only one word: True or False.\n"
        f"Answer:"
    )

    # Eingabe vorbereiten (auf GPU, wenn möglich)
    inputs = deepseek_tokenizer(prompt, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")

    # Antwort generieren
    outputs = deepseek_model.generate(**inputs, max_new_tokens=1)
    print("DeepSeek Output:", deepseek_tokenizer.decode(outputs[0], skip_special_tokens=True))

except Exception as e:
    print("DeepSeek Modell nicht geladen:", e)
```

## T5

```python
# 4. T5 (Encoder-Decoder, Klassifikation)
t5_tokenizer = T5Tokenizer.from_pretrained("t5-base")
t5_model = T5ForConditionalGeneration.from_pretrained("t5-base")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
t5_model = t5_model.to(device)

input_text = (
    f"Statement: {text}\n"
    f"Is this statement true or false?\n"
    f"Answer with only one word: True or False.\n"
    f"Answer:"
)

inputs = t5_tokenizer(input_text, return_tensors="pt").to(device)
outputs = t5_model.generate(**inputs, max_new_tokens=5)
label_t5 = t5_tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

print("T5 Klassifikation:", label_t5)
```

```python
# Lade FLAN-T5 statt T5
model_name = "google/flan-t5-base"
t5_tokenizer = T5Tokenizer.from_pretrained(model_name)
t5_model = T5ForConditionalGeneration.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
t5_model = t5_model.to(device)


# Prompt für Klassifikation
input_text = (
    f"Statement: {text}\n"
    f"Is this statement true or false?\n"
    f"Answer with only one word: True or False.\n"
    f"Answer:"
)

# Tokenisierung und Generierung
inputs = t5_tokenizer(input_text, return_tensors="pt").to(device)
outputs = t5_model.generate(**inputs, max_new_tokens=5)
label_t5 = t5_tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

print("FLAN-T5 Klassifikation:", label_t5)
```

## Bert

```python
# 5. BERT (Encoder-only, Klassifikation)
from transformers import BertTokenizer, BertForSequenceClassification
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
# WICHTIG: Hier nutzt du das Basis BERT-Modell ohne fine-tuning,
# das ist nur Demo und nicht gut für Fake-News Klassifikation!
bert_model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

bert_model = bert_model.to(device)
inputs = bert_tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)

with torch.no_grad():
    logits = bert_model(**inputs).logits
    predicted_class_id = torch.argmax(logits, dim=-1).item()

label_bert = "True" if predicted_class_id == 1 else "False"
print("BERT Output:", label_bert)
```

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification

# Modellname und Tokenizer laden
model_name = "roberta-base"
roberta_tokenizer = RobertaTokenizer.from_pretrained(model_name)
roberta_model = RobertaForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Gerät wählen (GPU oder CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
roberta_model = roberta_model.to(device)

# Eingabetext
text = "The Earth revolves around the Sun."

# Tokenisierung
inputs = roberta_tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)

# Vorhersage
with torch.no_grad():
    logits = roberta_model(**inputs).logits
    predicted_class_id = torch.argmax(logits, dim=-1).item()

label_roberta = "True" if predicted_class_id == 1 else "False"
print("RoBERTa Output:", label_roberta)
```

```python
import torch
import pandas as pd
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    AutoModelForSeq2SeqLM,
    RobertaTokenizer,
    RobertaForSequenceClassification,
    BitsAndBytesConfig,
)

# CUDA + bitsandbytes
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
bnb_config = BitsAndBytesConfig(load_in_4bit=True)

# Dein Prompt-Template
def build_prompt(text):
    return (
        f"Statement: {text}\n"
        f"Is this statement true or false?\n"
        f"Answer with only one word: True or False.\n"
        f"Answer:"
    )

# Liste der Modelle mit ihrer Architektur
models = {
    #"meta-llama/Llama-3.1-8B-Instruct": "causal",
    "EleutherAI/gpt-neo-2.7B": "causal"
}

# Daten laden (du kannst auch read_csv o. Ä. verwenden)
df = pd.read_csv("Datasets/LIAR/train_cleaned.csv", sep=",")  # ersetze mit deinem Pfad
df = df.iloc[:2000].copy()
statements = df["statement"].tolist()

# Ergebnisse speichern
results = {name: [] for name in models}

# Modelle durchlaufen
for model_name, model_type in models.items():
    print(f"Lade Modell: {model_name} ({model_type})")
    
    if model_type == "causal":
        tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
        model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
        
        for text in statements:
            prompt = build_prompt(text)
            inputs = tokenizer(prompt, return_tensors="pt").to(device)
            with torch.no_grad():
                output = model.generate(**inputs, max_new_tokens=5)
            decoded = tokenizer.decode(output[0], skip_special_tokens=True)
            result = decoded.split("Answer:")[-1].strip().split()[0]
            results[model_name].append(result)

    elif model_type == "seq2seq":
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
        
        for text in statements:
            prompt = build_prompt(text)
            inputs = tokenizer(prompt, return_tensors="pt").to(device)
            with torch.no_grad():
                output = model.generate(**inputs, max_new_tokens=5)
            decoded = tokenizer.decode(output[0], skip_special_tokens=True)
            results[model_name].append(decoded.strip())

    elif model_type == "roberta":
        tokenizer = RobertaTokenizer.from_pretrained(model_name)
        model = RobertaForSequenceClassification.from_pretrained(model_name, num_labels=2).to(device)
        
        for text in statements:
            inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
            with torch.no_grad():
                logits = model(**inputs).logits
                pred = torch.argmax(logits, dim=-1).item()
            results[model_name].append("True" if pred == 1 else "False")

# Ergebnisse in DataFrame speichern
for model_name in results:
    df[f"label_{model_name.split('/')[-1]}"] = results[model_name]

# Export
df.to_csv("statements_with_predictions_6.csv", index=False)
print("Fertig – Ergebnisse gespeichert in 'statements_with_predictions.csv'")
```

```python
import pandas as pd

# Dateinamen deiner CSVs
files = [
    "statements_with_predictions_2.csv",
    "statements_with_predictions_3.csv",
    "statements_with_predictions_4.csv",
    "statements_with_predictions_5.csv",
    "statements_with_predictions.csv",
    "statements_with_predictions_6.csv"
]

# Die ersten drei Spaltennamen (ersetze ggf. durch deine echten Namen)
first_cols = ["label", "statement", "subjects"]

# Lade die erste Datei komplett
df_merged = pd.read_csv(files[0])

# Füge die restlichen Dateien hinzu (ohne die ersten 3 Spalten)
for file in files[1:]:
    df = pd.read_csv(file)
    df = df.drop(columns=first_cols)  # entferne die ersten 3 Spalten
    df_merged = pd.concat([df_merged, df], axis=1)

# Speichere das Ergebnis
df_merged.drop('label_gpt2', axis=1, inplace=True)
df_merged.to_csv("merged.csv", index=False)
df_merged
```

```python
import pandas as pd

# Lade die Datei mit den Original-Labels
df_merged = pd.read_csv("merged.csv")

# Mapping: Original-Label zu "True"/"False"
true_labels = ["true", "mostly-true", "half-true"]
false_labels = ["barely-true", "false", "pants-fire", "pants-on-fire"]

def map_label(label):
    label = str(label).strip().lower()
    if label in true_labels:
        return "True"
    elif label in false_labels:
        return "False"
    else:
        return "False"  # Standardwert

# Neue Spalte mit binärem Label
df_merged["label_binary"] = df_merged["label"].apply(map_label)

# Optional: Speichern
df_merged.to_csv("merged_with_binary_label.csv", index=False)
df_merged
```

```python
import pandas as pd
import numpy as np

# Lade das DataFrame
df = pd.read_csv("merged_with_binary_label.csv")

# Modellspalten
model_cols = [
    "label_flan-t5-base",
    "label_deepseek-llm-7b-base",
    "label_gemma-2b-it",
    "label_roberta-base",
    "label_Llama-3.1-8B-Instruct",
    "label_gpt-neo-2.7B"
]

# Bereinigen: Nur "True" oder "False", alles andere zu NaN
for col in model_cols:
    # Entferne Punkt und Leerzeichen, mache Großbuchstaben
    df[col] = df[col].astype(str).str.strip().str.replace(".", "", regex=False).str.capitalize()
    # Setze alles, was nicht exakt "True" oder "False" ist, auf NaN
    df[col] = df[col].where(df[col].isin(["True", "False"]), np.nan)

# Optional: Speichern
df.to_csv("merged_with_binary_label_cleaned.csv", index=False)
df.head()
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Lade die Datei mit den binären Labels
df = pd.read_csv("merged_with_binary_label_cleaned.csv")

# Liste der Modellspalten (anpassen, falls andere Namen)
model_cols = [
    "label_flan-t5-base",
    "label_deepseek-llm-7b-base",
    "label_gemma-2b-it",
    "label_roberta-base",
    "label_Llama-3.1-8B-Instruct",
    "label_gpt-neo-2.7B"
]

# Ground Truth
y_true = df["label_binary"].astype(str).str.capitalize()

for col in model_cols:
    # Predictions (ggf. auf True/False normalisieren)
    y_pred = df[col].astype(str).str.strip().str.capitalize().replace({"False.": "False", "True.": "True"})
    cm = confusion_matrix(y_true, y_pred, labels=["True", "False"])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["True", "False"])
    disp.plot()
    plt.title(f"Confusion Matrix: {col}")
    plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

for col in model_cols:
    y_pred = df[col].astype(str).str.strip().str.capitalize()
    print(f"--- {col} ---")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, average="macro", pos_label=None))
    print("Recall:", recall_score(y_true, y_pred, average="macro"))
    print("F1-Score:", f1_score(y_true, y_pred, average="macro"))
    print(classification_report(y_true, y_pred, labels=["True", "False"]))
```

# Execution with subjects

```python
import torch
import pandas as pd
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    AutoModelForSeq2SeqLM,
    RobertaTokenizer,
    RobertaForSequenceClassification,
    BitsAndBytesConfig,
)

# CUDA + bitsandbytes
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
bnb_config = BitsAndBytesConfig(load_in_4bit=True)

# Prompt-Template mit Statement + Subjects
def build_prompt(statement, subjects):
    return (
        f"Subjects: {subjects}\n"
        f"Statement: {statement}\n"
        f"Is this statement true or false?\n"
        f"Answer with only one word: True or False.\n"
        f"Answer:"
    )

# Liste der Modelle mit ihrer Architektur
models = {
    #"meta-llama/Llama-3.1-8B-Instruct": "causal",
    "EleutherAI/gpt-neo-2.7B": "causal"
}

# Daten laden
df = pd.read_csv("Datasets/LIAR/train_cleaned.csv", sep=",")
df = df.iloc[:2000].copy()

statements = df["statement"].tolist()
subjects = df["subjects"].tolist()

# Ergebnisse Container
results = {name: [] for name in models}

# Modelle durchlaufen
for model_name, model_type in models.items():
    print(f"Lade Modell: {model_name} ({model_type})")

    if model_type == "causal":
        tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
        model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")

        for text, subj in zip(statements, subjects):
            prompt = build_prompt(text, subj)
            inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(device)
            with torch.no_grad():
                output = model.generate(**inputs, max_new_tokens=5)
            decoded = tokenizer.decode(output[0], skip_special_tokens=True)
            result = decoded.split("Answer:")[-1].strip().split()[0]
            results[model_name].append(result)

    elif model_type == "seq2seq":
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")

        for text, subj in zip(statements, subjects):
            prompt = build_prompt(text, subj)
            inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(device)
            with torch.no_grad():
                output = model.generate(**inputs, max_new_tokens=5)
            decoded = tokenizer.decode(output[0], skip_special_tokens=True)
            results[model_name].append(decoded.strip())

    elif model_type == "roberta":
        tokenizer = RobertaTokenizer.from_pretrained(model_name)
        model = RobertaForSequenceClassification.from_pretrained(model_name, num_labels=2).to(device)

        for text in statements:  # Roberta verwendet nur das statement
            inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
            with torch.no_grad():
                logits = model(**inputs).logits
                pred = torch.argmax(logits, dim=-1).item()
            results[model_name].append("True" if pred == 1 else "False")

# Ergebnisse im DataFrame speichern
for model_name in results:
    df[f"label_{model_name.split('/')[-1]}"] = results[model_name]

# Export
output_file = "statements_with_predictions_subjects1.csv"
df.to_csv(output_file, index=False)
print(f"✅ Fertig – Ergebnisse gespeichert in '{output_file}'")
```
